```python
import numpy
import cv2
import imutils
import os
```

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('object.jpg')
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(20,8))
plt.imshow(gray)
```

```python
edged = cv2.Canny(gray, 40, 150)
plt.figure(figsize=(20,8))
plt.imshow(edged)
```

```python
thresh = cv2.threshold(gray, 225,255, cv2.THRESH_BINARY_INV)[1]
plt.figure(figsize=(20,8))
plt.imshow(thresh)
```

```python
cont = cv2.findContours(thresh.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
cont = imutils.grab_contours(cont)
output = img.copy()

for c in cont:
  cv2.drawContours(output, [c],-1, (0,0,255), 3)
plt.figure(figsize=(20,8))
plt.imshow(output)
```

```python
text = "{} objects found".format(len(cont))
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org = (50, 50) 
  
# fontScale 
fontScale = 1
   
# Blue color in BGR 
color = (255, 0, 0) 
  
# Line thickness of 2 px 
thickness = 2
   
# Using cv2.putText() method 
image = cv2.putText(output, text, org, font,  
                   fontScale, color, thickness, cv2.LINE_AA)
```

```python
plt.figure(figsize=(20,8))
plt.imshow(image)
```
